```python
import pandas as pd

# Read in the dataset
df = pd.read_excel('rawdata.xlsx')

# Calculate the mean of the relevant columns
means = df[["B1SD", "B1SND", "B1EAD", "B1EAND", "SHD", "SHND", "S1D", "S1ND", "RHD", "RHND", "RFD", "RFND"]].mean(axis=1)

# Create a new column called "Health status"
df["Health status"] = 1  # initialize all values to 1
df.loc[means < 350, "Health status"] = 0  # set values to 0 where mean is less than 50

# Print the updated dataframe
print(df)
```

```python
# count the number of 0s and 1s in the "Health status" column
df["Health status"].value_counts()
```

```python
# save the dataframe to a new csv file without the index column
df.to_csv("data.csv", index=False)
```
